```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('..\\ny_crime_13.xls', header=4)
df.head()
```

```python
df_features = pd.DataFrame()
df_features['pop'] = (df.Population - df.Population.mean())/df.Population.std()
df_features['pop_squared'] = df_features.pop**2
df_features['murder'] = np.where(df.iloc[:, 3] > 0, 1, 0)
df_features['robbery'] = np.where(df.iloc[:, 6] > 0, 1, 0)
df_features.head()
```

```python
# Restructure the data so we can use facetgrid
df_long = df_features
df_long = pd.melt(df_long, id_vars=['murder', 'robbery'])
g = sns.FacetGrid(df_long, col='variable', size = 5, aspect=.7)
g = g.map(sns.boxplot, 'murder', 'value')
plt.tight_layout()
plt.show()
```
